```python
#Question 4
n=int(input())
i,x,y=0,0,0
while(i<n):
    d,v=input().split()
    if(d[0]=="U"):
        x+=int(v)
    elif(d[0]=="D"):
        x-=int(v)
    elif(d[0]=="R"):
        y+=int(v)
    elif(d[0]=="L"):
        y-=int(v)
    i+=1
print(round(((x**2)+(y**2))*0.5))
```
